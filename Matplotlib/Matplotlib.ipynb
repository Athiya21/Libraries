- This project is all about Matplotlib, the basic visualization tool of python programming language

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
#here 0 -> start
#10 -> stop
#100 -> space
x1 = np.linspace(0,10,100) #Return evenly spaced numbers over a specified interval
fig = plt.figure()
plt.plot(x1,np.sin(x1),'-')
plt.plot(x1,np.cos(x1),'--')
```

```python
plt.figure()
#create the first of two panels & set current axis
plt.subplot(2,1,1)
plt.plot(x1,np.sin(x1))
```

```python
plt.figure()
#create the first of two panels & set current axis
plt.subplot(2,1,1) #rows, columns, panel number(or) plot
plt.plot(x1,np.sin(x1))
#create the second of two panels & set current axis
plt.subplot(2,1,2)
plt.plot(x1,np.cos(x1))
```

## Object-Oriented API

```python
#First create a grid of plots
#ax will be an array of two axes object
fig, ax = plt.subplots(2)
```

```python
#First create a grid of plots
#ax will be an array of two axes object
fig, ax = plt.subplots(2)

#call plot() method on the appropriate object
ax[0].plot(x1,np.sin(x1),'b-') # b => blue line   - => solid line style
ax[1].plot(x1,np.cos(x1),'b-')
```

## Objects and Reference

```python
fig = plt.figure()
x2 = np.linspace(0,5,10)
y2 = x2 ** 2
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x2,y2,'r')
```

```python
fig = plt.figure()
x2 = np.linspace(0,5,10)
y2 = x2 ** 2
axes = fig.add_axes([0.1,0.1,0.8,0.8]) #[left, bottom, width, height]
axes.plot(x2,y2,'r')
axes.set_xlabel('x2')
axes.set_ylabel('y2')
axes.set_title('title')
```
