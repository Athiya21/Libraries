- displot -> distribution plot
- plot the graphi using 1 variable is known as univariate analysis

```python
vis1 = sns.distplot(df["InternetUsers"]) #univariate analysis statistics
```

```python
vis2 = sns.distplot(df["InternetUsers"],bins = 10)
```

```python
vis3 = sns.distplot(df["InternetUsers"],bins = 20)
```

- Box plots

```python
vis4 = sns.boxplot(data = df,x = "IncomeGroup",y = "BirthRate") #Bivariate Analysis
```

- lmplot -> linear model plot

```python
vis5 = sns.lmplot(data=df, x ='InternetUsers', y = 'BirthRate')
```

```python
vis6 = sns.lmplot(data = df, x = 'InternetUsers', y = 'BirthRate', fit_reg = False)
```

```python
vis7 = sns.lmplot(data = df, x = 'InternetUsers', y = 'BirthRate', fit_reg = True)
```

- hue => parameter of the color

```python
vis8 = sns.lmplot(data = df, x = 'InternetUsers', y = 'BirthRate', fit_reg = True, hue = 'IncomeGroup')
```

```python
vis9 = sns.lmplot(data = df, x = 'InternetUsers', y = 'BirthRate', fit_reg = True, hue = 'IncomeGroup', size = 10)
```
